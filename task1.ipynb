```python
import plotly.express as px
rel_dur_rat = px.bar(df, x = 'Cocoa\nPercent' ,color='Cocoa\nPercent')
rel_dur_rat.update_layout(title='cocoa content in percent',
                          title_x=0.5,
                          title_pad=dict(t=20),
                          title_font=dict(size=20),
                          plot_bgcolor = 'white')
```

```python
import plotly.express as px
rel_dur_rat = px.histogram(df, x = 'Rating' ,color='Rating')
rel_dur_rat.update_layout(title='rating of bar chocolates',
                          title_x=0.5,
                          title_pad=dict(t=20),
                          title_font=dict(size=20),
                          plot_bgcolor = 'white')
```
